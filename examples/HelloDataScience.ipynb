```python
import pandas as pd
from sodapy import Socrata
import matplotlib.pyplot as plt

# Request the library dataset 
client = Socrata("data.sfgov.org", None)
results = client.get("qzz6-2jup", limit=4000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
fig, ax = plt.subplots()

# Make the bar chart
ax.bar(
    results_df['patron_type_definition'], 
    results_df['total_checkouts'].astype('int32'), # You need to convert the "total checkouts" to int because pandas registered it as a string
)

# Set the x labels to be vertical so they all fit
plt.setp(ax.get_xticklabels(), fontsize=10, rotation='vertical')

# Show the plot
plt.show()
```
