```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm

##Let's create a chart with some arbitrary numbers
plt.plot([1, 2, 3, 4],[5, 6, 7, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
#Let's take a look at our data
endpoint = "https://data.austintexas.gov/resource/wter-evkm.csv"
df = pd.read_csv(endpoint, sep=',')
df.head()
```

```python
##Let's sort by a category
animals = df.groupby(['animal_type']).size()
animals.head()
animals.plot.bar(x='animal_type')
```

```python
##Let's investigate this "other" type
others = df[df['animal_type']=='Other']
others.head()
others_breed = others.groupby(['breed']).size().sort_values(ascending=False)
```

```python
##Create a plot and make it more legibile 

color = cm.inferno_r(np.linspace(.4, .8, 30))
others_plot = others_breed.plot.bar(x='breed', title="Austin found animals that are not cats, dogs, or birds", color=color)
others_plot.set_xlabel("Breed")
others_plot.set_ylabel("Number found")
```

```python
##Another color option

color = ['g','b','y','k']*4
others_plot = others_breed.plot.bar(x='breed', title="Austin found animals that are not cats, dogs, or birds", color=color)
others_plot.set_xlabel("Breed")
others_plot.set_ylabel("Number found")
```
